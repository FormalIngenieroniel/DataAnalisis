```python
import sqlite3
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline

conn = sqlite3.connect('/content/InternetSpeed.db')

df = pd.read_csv('/content/LA_population.csv')
df2 = pd.read_csv('/content/LA_wifi_speed_UK.csv')

df.to_sql('LA_population',conn,if_exists = 'replace', index = False)
df2.to_sql('LA_wifi_speed_UK',conn,if_exists = 'replace', index = False)

cur = conn.cursor()

query = 'SELECT * FROM LA_wifi_speed_UK LIMIT 10;'

cur.execute(query)

for row in cur:
  print(row)

cur.description[:10]

columns = [member[0] for member in cur.description]

columns = columns[1:]

columns = [c.replace('_p','') for c in columns]
columns = [c.replace('_d','') for c in columns]
columns = [c.replace('_u','') for c in columns]

columns = list(set(columns))
columns[:10]
```
