```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
```

```python
salesDit = pd.read_csv('stores-dist.csv')
salesDit.columns = ['district','sales','stores']

salesDit.head()
```

```python
sales = salesDit.drop('district', axis=1)
sales.head()
```

```python
y = sales.sales
x = sales.stores
```

```python
order = 1
p = np.poly1d(np.polyfit(x,y,order))

print('p(x) tiene el valor polinomial de cada y para cada x,\n\n{}.'.format(p(x)))
print('\nEl vector de coeficientes p describe el modelo de regresion:\n{}'.format(p))
print('\nEl p[0]: {}.'.format(p[0]))
print('\nEl p[1]: {}.'.format(p[1]))
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y,p(x))
r2
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y,p(x))
mse
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y,p(x))
mae
```
