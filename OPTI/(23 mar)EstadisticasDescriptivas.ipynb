```python
datos['Rounded'] = datos['weight'].round(2)
datos.head(5)
```

```python
datos['Diff'] = datos['Rounded'] - 20
datos.describe()
```

```python
count = datos['Rounded'].count()
mean = datos['Rounded'].mean()
median = datos['Rounded'].median()
std = datos['Rounded'].std()
min = datos['Rounded'].min()
max = datos['Rounded'].max()
rng = datos['Rounded'].max() - datos['Rounded'].min()
```

```python
 countstring = 'La cuenta (count) es: {}'.format(count)
 meanstring = 'El media (mean) es: {}'.format(mean)
 stdstring = 'La desviacion standart (std) es: {}'.format(std)
 rangestring = 'El rango (range) es: {}'.format(rng)
 print(countstring)
 print(meanstring)
 print(stdstring)
 print(rangestring)
```

```python
count = datos.Rounded.count()
stdstring = 'The standard deviation of the distribution is {:.2f}.'.format(std)
print(count)
print(stdstring)
```

```python
freq = datos['Rounded'].value_counts().to_frame().reset_index()
print(type(freq))
freq.head(5)
```

```python
freq.columns = ['value','freq']
freq.head(5)
```

```python
plt.figure(figsize=(20,10))
plt.ylabel('Frequency')
plt.xlabel('Weight')
plt.plot(freq.value,freq.freq, "o", markersize = 10, color = 'g')
```
