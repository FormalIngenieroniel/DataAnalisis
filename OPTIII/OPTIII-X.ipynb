# **Metricas de regresion:**

mse -> [0,inf)

rmse -> (-inf,inf)

mae -> [0,inf)

r^2 -> [0,1]

En el caso del r cuadrado necesitamos que nos de cercan a uno pero si da menor a cero quiere indicar que es peor que un promedio.

Las metricas serviran cuando los datos no estan balanceados

**ACCURACY:** Se evaluan ambas ramas de clasificacion, se mira cuantos se predijeron correctamente en los positivos

**PRECISION:** Cual es la probabilidad de los que se predijeron positivos (rama de prediccion positiva)

**RECALL:** Cuantos positivos se clasificaron bien, se ve ambas ramas (rama de prediccion positiva)
