```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
dflist = [
    {'edad': 22, 'sexo': 1, 'cliente': "Bueno"},
    {'edad': 30, 'sexo': 0, 'cliente': "Malo"},
    {'edad': 35, 'sexo': 1, 'cliente': "Bueno"},
    {'edad': 21, 'sexo': 0, 'cliente': "Bueno"},
    {'edad': 40, 'sexo': 1, 'cliente': "Bueno"},
    {'edad': 42, 'sexo': 1, 'cliente': "Bueno"}
]

df = pd.DataFrame(dflist)
df
```

```python
model = DecisionTreeClassifier()
#Cuando se cambia el argumento va cambiando los datos, buscamos que ambos resultados sean parecidos.
model.fit(df[["edad","sexo"]],df["cliente"])
```

```python
from sklearn.tree import plot_tree
plot_tree(model, feature_names=["edad","sexo","Chol"])
```

```python
imp= model.feature_importances_
```

```python
plt.figure(figsize=(8,5))
plt.bar(['edad','sexo'], imp, color='blue')
```
