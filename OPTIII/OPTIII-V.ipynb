```python
import scipy
import numpy as np
import seaborn as sns

poblacion = np.random.normal(5, 2,10)

sns.histplot(poblacion)

scipy.stats.norm.cdf(2,5,2)
```

```python
poblacion = np.random.normal(5,4,10000)
#Se crea la poblacion con 1k datos de donde se sacaran las muestras (en este caso se sacan 10)

means = []
for i in range(1000):
  sample = np.random.choice(poblacion, size=5)
  means.append(np.mean(sample))

sns.histplot(means)

mean_of_means = np.mean(means)
print(f"The mean of the means is: {mean_of_means}")

std_of_means = np.std(means)
print(f"The std of the means is: {std_of_means}")

scipy.stats.norm.cdf(2.5,mean_of_means,std_of_means)
```

```python
scipy.stats.norm.cdf(2,mean_of_means,std_of_means)
```
