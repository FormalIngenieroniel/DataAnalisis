```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/Advertising.csv')
df

#La variable que se va a predecir estara en el eje Y, en caso del ejemplo es la variable sales
#Se hara la prediccion con las columnas TV, radio y newspaper. Estas variables van en el eje x (se llaman predictores o caracteristicas)
#se ven dos tipos de modelos de regresion, donde las variables son continuas (numericas) o si son de clasificacion (categoricas)

plt.scatter(df['TV'], df['sales'], label='TV')

# Add a legend to the plot
plt.legend()

# Add a title to the plot
plt.title('Scatter Plot of Sales vs. Tv')

# Show the plot
plt.show()
```

```python
# Create another scatter plot with sales as the y-axis and radio as the x-axis
plt.scatter(df['radio'], df['sales'], label='Radio')
# Add a legend to the plot
plt.legend()

# Add a title to the plot
plt.title('Scatter Plot of Sales vs. Radio')

# Show the plot
plt.show()
```

```python
# Create another scatter plot with sales as the y-axis and newspaper as the x-axis
plt.scatter(df['newspaper'], df['sales'], label='Newspaper')

# Add a legend to the plot
plt.legend()

# Add a title to the plot
plt.title('Scatter Plot of Sales vs. Newspaper')
plt.show()
```

```python
sns.histplot(df['sales'], bins=10, kde=True)
```
