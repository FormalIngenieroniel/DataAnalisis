- ¿qué significa el valor del punto de corte para el problema que se soluciona?

El valor del punto de corte es muy importante pues es este el que indica el valor base de la variable dependiente cuando no hay influencia de las variables independientes (Como si las otras fueran 0).

En este ejemplo, en un modelo de regresión que predice las ventas basado en tv, pariodico y la radio. El punto de corte podría representar el valor de las ventas con caro inversion en publicidad en los medios como tv, radio y periodico.

- ¿qué significa el valor de cada coeficiente  para el problema que se soluciona? cómo influyen en la variable a predecir?

Los coeficientes en un modelo de regresión lineal tienen la funcion de mostrar la magnitud y dirección de una relación entre cada variable independiente y la variable dependiente. Cada coeficiente indica cuánto cambia la variable dependiente cuando la variable independiente determinada se incrementa por 1, mientras que todas las demás variables se dejan constantes.

En este ejemplo, si en el mismo modelo de regresión de ventas, el coeficiente para tv es 0.04663016, significa que, en promedio, por cada unidad adicional de inversion en tv, el valor de las ventas aumenta en 0.04663016, manteniendo constantes otras variables como el periodico y radio.

Por lo tanto, el valor de cada coeficiente es crucial para entender cómo influyen las variables independientes en la variable dependiente y cómo se puede usar el modelo para hacer predicciones sobre la variable dependiente en función de los valores de las variables independientes.
