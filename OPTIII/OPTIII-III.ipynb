```python
import numpy as np
import seaborn as sns

np.random.seed(123)

poblacion = np.random.normal(172, 10,10000)

sns.histplot(poblacion)
```

```python
# prompt: how many data are between 162 and 182

import numpy as np

# Calculate the number of data points between 162 and 182
num_data_between = np.sum((poblacion >= 162) & (poblacion <= 182))

# Print the result
print(f"La cantidad de manes between 162 and 182 es: {num_data_between}")
```

```python
# prompt: generate the median and std of poblacion

import numpy as np
median = np.median(poblacion)
std = np.std(poblacion)

print(f"The median of the population is: {median}")
print(f"The standard deviation of the population is: {std}")
```

```python
# prompt: Create a for with 1000 iterations that print the different means with the line sample = np.random.choice(poblacion, size = 100) and take the distribution of the 100 means and take the mena of that distribution

import numpy as np
means = []
for i in range(1000):
  sample = np.random.choice(poblacion, size=100)
  means.append(np.mean(sample))

import seaborn as sns
sns.histplot(means)

mean_of_means = np.mean(means)
print(f"The mean of the means is: {mean_of_means}")

std_of_means = np.std(means)
print(f"The std of the means is: {std_of_means}")
```
