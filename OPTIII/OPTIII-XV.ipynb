METODOS DE REFERENCIA PARA SABER SI QUE LOS QUE TENEMOS SON MEJORES (NAIVE O PERSISTENCIA Y MEDIA MOVIL)

Persistencia: El de hoy es el mismo de mañana

Media movil: el promedio de los n dias anteriores

```python
import pandas as pd

#Parse dates se encarga de ponerle un formato al indice
df = pd.read_csv('/content/USDCOP=X.csv', index_col=0, parse_dates=True)

df.info()
```

```python
ts = df['Close']
train = ts[:'2024-05-01']
test= ts['2024-05-01':]
```

```python
df_train = pd.DataFrame(train.values, index = train.index, columns = ['t'])
for i in range (1,4):
  df_train[f't-{i}'] = df_train['t'].shift(i)
df_train = df_train.dropna()
df_train
```

```python
y_train = df_train['t']
X_train = df_train.drop('t', axis=1)
```

```python
# prompt: Saca el error cuadratico medio de train

import numpy as np
from sklearn.metrics import mean_squared_error

y_pred = df_train['t-1']
mse = mean_squared_error(y_train, y_pred)

print("Train RMSE:", np.sqrt(mse))
```

```python
import numpy as np
df_test = pd.DataFrame(test.values, index = test.index, columns = ['t'])
for i in range (1,4):
  df_test[f't-{i}'] = df_test['t'].shift(i)
df_test = df_test.dropna()
y_test = df_test['t']
X_test = df_test.drop('t', axis=1)

y_pred_test = df_test['t-1']
mse_test = mean_squared_error(y_test, y_pred_test)

print("Test RMSE:", np.sqrt(mse_test))
```

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor

# Create a decision tree regressor
model = DecisionTreeRegressor(max_depth=6)

# Train the model
model.fit(X_train, y_train)

# Make predictions on the train and test sets
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Calculate the mean squared error for train and test sets
mse_train = mean_squared_error(y_train, y_pred_train)
mse_test = mean_squared_error(y_test, y_pred_test)

# Print the train and test RMSE
print("Train RMSE:", np.sqrt(mse_train))
print("Test RMSE:", np.sqrt(mse_test))
```

```python
from sklearn.ensemble import RandomForestRegressor

# Create a random forest regressor
model = RandomForestRegressor(max_depth=5, n_estimators=10)

# Train the model
model.fit(X_train, y_train)

# Make predictions on the train and test sets
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Calculate the mean squared error for train and test sets
mse_train = mean_squared_error(y_train, y_pred_train)
mse_test = mean_squared_error(y_test, y_pred_test)

# Print the train and test RMSE
print("Train RMSE:", np.sqrt(mse_train))
print("Test RMSE:", np.sqrt(mse_test))
```

```python
from sklearn.neighbors import KNeighborsRegressor

# Create a KNN regressor
model = KNeighborsRegressor(n_neighbors=2)

# Train the model
model.fit(X_train, y_train)

# Make predictions on the train and test sets
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Calculate the mean squared error for train and test sets
mse_train = mean_squared_error(y_train, y_pred_train)
mse_test = mean_squared_error(y_test, y_pred_test)

# Print the train and test RMSE
print("Train RMSE:", np.sqrt(mse_train))
print("Test RMSE:", np.sqrt(mse_test))
```

```python
from sklearn.linear_model import LinearRegression

# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the train and test sets
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)

# Calculate the mean squared error for train and test sets
mse_train = mean_squared_error(y_train, y_pred_train)
mse_test = mean_squared_error(y_test, y_pred_test)

# Print the train and test RMSE
print("Train RMSE:", np.sqrt(mse_train))
print("Test RMSE:", np.sqrt(mse_test))
```

```python
# prompt: Implementa el modelo naive con el rmse en train y test

import numpy as np
# Naive method: predict the next value as the same as the current value
y_pred_naive_train = X_train['t-1']
y_pred_naive_test = X_test['t-1']

# Calculate the mean squared error for train and test sets
mse_naive_train = mean_squared_error(y_train, y_pred_naive_train)
mse_naive_test = mean_squared_error(y_test, y_pred_naive_test)

# Print the train and test RMSE
print("Naive Train RMSE:", np.sqrt(mse_naive_train))
print("Naive Test RMSE:", np.sqrt(mse_naive_test))
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error

# Moving average method: predict the next value as the average of the last n values
n = 3

# Calculate the moving average for train and test sets
y_pred_ma_train = df_train['t'].rolling(window=n).mean().shift(1)
y_pred_ma_test = df_test['t'].rolling(window=n).mean().shift(1)

# Remove NaN values from predictions and true values
y_pred_ma_train = y_pred_ma_train.dropna()
y_pred_ma_test = y_pred_ma_test.dropna()
y_train = y_train[y_train.index.isin(y_pred_ma_train.index)]
y_test = y_test[y_test.index.isin(y_pred_ma_test.index)]

# Calculate the mean squared error for train and test sets
mse_ma_train = mean_squared_error(y_train, y_pred_ma_train)
mse_ma_test = mean_squared_error(y_test, y_pred_ma_test)

# Print the train and test RMSE
print("Moving Average Train RMSE:", np.sqrt(mse_ma_train))
print("Moving Average Test RMSE:", np.sqrt(mse_ma_test))
```
