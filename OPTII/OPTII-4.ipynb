```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import spearmanr
```

```python
np.random.seed(123)
x = np.arange(0,5,0.1)
```

```python
#Para pearson 0.6, se varia con 5
#Para pearson 0, se varia con numeros muy grandes
#Para pearson -0.6, se varia con pendiente negativa y un 1

y= (2 * x*x + 3 + np.random.normal(0,0,50))
```

```python
plt.scatter(x,y)
```

```python
#matriz de correlacion de x y y (incluyen a si mismas)
#es mas fuerte cuando toma un comportamiento lineal

np.corrcoef(x,y)
```

```python
spearmanr(x,y)
#cuando sospecho corr pero no es lineal
```

```python
df = pd.DataFrame({"x": x,"y":y})
```

```python
df[['x','y']].corr()
```

```python
df[['x','y']].corr(method='spearman')
```
