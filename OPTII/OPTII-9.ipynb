```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from sklearn.impute import KNNImputer
```

```python
np.random.seed(123)
a = np.random.uniform(0,5,(10,4))
veci = 2

a[2,3] = np.nan
a[9,1] = np.nan
a[5,3] = np.nan
a[7,2] = np.nan
```

```python
df = pd.DataFrame(data=a, columns=["x1","x2","x3","x4"])
df
```

```python
for index, row in df.iterrows():
  print(row['x1'])
```

```python
dicc = {}
for a in range(0,10):
    c1 = df.iloc[a,0] - 2.404660
    c2 = df.iloc[a,1] - 1.960588
    c3 = df.iloc[a,2] - 1.715890
    h = np.sqrt(c1**2+c2**2+c3**2)
    print(h)
    if ~np.isnan(h):
      dicc.update({a:h})
```

```python
print(dicc)
```

```python
imputer = KNNImputer(n_neighbors=2)

imputer.fit(df)
imputer.transform(df)
```
