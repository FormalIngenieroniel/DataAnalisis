```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
import joblib
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/camilousa/datasets/master/german_credit_data.csv')
```

```python
df.info()
```

```python
df
```

```python
imputer1 = SimpleImputer(strategy='median')
y1 = StandardScaler()

pipelineNum =\
  Pipeline(
      [
          ('operacion1', imputer1),('operacion2', y1),
      ]
  )

imputer2 = SimpleImputer(strategy='most_frequent')
y2 = OneHotEncoder()

pipelineCat =\
  Pipeline(
      [
          ('operacion3', imputer2),('operacion4', y2),
      ]
  )
```

```python
transformer = ColumnTransformer(
    [
        ('PipelineNum',pipelineNum,[0,1,3,7,8]),
        ('PipelineCat',pipelineCat,[2,4,5,6,9,10])
    ]
)
transformer.fit(df)
```

```python
joblib.dump(transformer,"model.joblib")
df.shape
```
