```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances
```

```python
df = pd.DataFrame({
    "NotaMaten_I":[2.8,5.0,4.9,2.1,3.0,4.1,2.1,1.5],
    "NotaMaten_2":[1.5,4.0,4.2,3.0,3.2,4.5,2.5,2.0],
    "Nota_algoritmos":[np.nan,4.0,2.1,4.0,np.nan,3.0,2.8,4.0]
})

df
```

```python
imputer = SimpleImputer()
#para utilizar mediana o asi se pone imputer = SimpleImputer(strategy='median)
imputer.fit(df)
#fit calcula la   estadistica para imputar (no hace la imputacion solo los calculos)
imputer.statistics_
#Trae la media de cada columna del df
imputer.transform(df)
#Imputa con la estrategia donde haga falta
```

```python
plt.scatter(df["NotaMaten_I"], df["NotaMaten_2"])
```

```python
c1 = 5.0 - 2.8
c2 = 4.0 - 1.5
h = np.sqrt(c1**2+c2**2)
h
print("Distancias desde 0")
for a in range(1,8):
  c1 = df.iloc[a,0] - 2.8
  c2 = df.iloc[a,1] - 1.5
  h = np.sqrt(c1**2+c2**2)
  print(h)
```

```python
print("Distancias desde 4")
for a in range(0,8):
  c1 = df.iloc[a,0] - 3.0
  c2 = df.iloc[a,1] - 3.2
  h = np.sqrt(c1**2+c2**2)
  print(h)
```
