```python
#Acitividad
Df = pd.read_csv('/content/USDCOP=X.csv')
Df = Df[{'Date','Close'}]
Df
```

```python
Ts = pd.Series(index=Df['Date'].values, data = Df['Close'].values)
Ts.index = pd.to_datetime(Ts.index)
Ts.plot()
```

```python
BS = (Ts.rolling(window=20).mean() + 2 * (Ts.rolling(window=20).std()))
BI = (Ts.rolling(window=20).mean() - 2 * (Ts.rolling(window=20).std()))

plt.plot(BS)
plt.plot(BI)
plt.plot(Ts)
plt.show()
```

```python
Comprar = Ts < BI
Vender = Ts > BS
```

```python
plt.plot(Ts[Comprar])
plt.show()
```

```python
plt.plot(Ts[Vender])
plt.show()
```
