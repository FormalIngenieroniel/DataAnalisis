```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
Df = pd.DataFrame(
    {
        "col1": [1,3, np.nan, 5],
        "col2": [5,1,2,5]
    }
)
Df.info()
```

```python
imputer = SimpleImputer(strategy='median')
y = MinMaxScaler()

pipeline =\
  Pipeline(
      [
          ('nombre', imputer),('nombre2', y),
      ]
  )

pipeline.fit(Df)
pipeline.transform(Df)
```

```python
pipeline.transform([[0,np.nan]])
```

```python
#trabajo en clase
DfII = pd.read_csv('/content/titanic.csv')
DfII.info()
```

```python
DfNum = DfII[{"Survived","Pclass","Age","SibSp","Parch","Fare"}]
DfCat = DfII[{"Sex","Ticket","Cabin","Embarked"}]
```

```python
imputer1 = SimpleImputer(strategy='median')
y1 = StandardScaler()

pipelineNum =\
  Pipeline(
      [
          ('operacion1', imputer1),('operacion2', y1),
      ]
  )
```

```python
imputer2 = SimpleImputer(strategy='most_frequent')
y2 = OneHotEncoder()

pipelineCat =\
  Pipeline(
      [
          ('operacion3', imputer2),('operacion4', y2),
      ]
  )
```

```python
transformer = ColumnTransformer(
    [
        ('PipelineNum',pipelineNum,[1,2,5,6,7,9]),
        ('PipelineCat',pipelineCat,[4,8,10,11])
    ]
)
transformer.fit(DfII)
transformer.transform(DfII)
```
