Onehold crea una columna por cada valor que hay, las columnas tendran valores de 0 a 1 dependiendo si existe el valor.

Con muchos valores el dataframe crece bastante.

Ordinal Asigna un numero a cada valor categorico, el numero sera <= 0, el numero sera unico.

Sirve para dar un orden a las variables, por ejemplo, si usamos jerarquias se puede plasmar con los numeros. Entre mayor el numero indica que mejor sera la categorica, solo se puede aplicar a este contexto cuando hablamos de modelos.

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
Df = pd.DataFrame(
    {"color": ["amarillo","rojo","amarillo","azul","amarillo","azul",],
     "tipo_cliente": ["bueno","malo","regular","bueno","bueno","regular"]}
)
```

```python
#sparse guarda false guarda todo, con true solo guarda la posicion del 1
encoder = OneHotEncoder(sparse=False)
encoder.fit(Df[["color"]])
encoder.transform(Df[["color"]])
#Hacer el analisis asi esta feo, es mejor hacer esto con el modelo
```

```python
#Para poder ver el nombre de las columnas
encoder.categories_
```

```python
#Matris de los arreglos de cada una columna
encoder = OrdinalEncoder(categories=[["malo","regular","bueno"]])
encoder.fit(Df[["tipo_cliente"]])
encoder.transform(Df[["tipo_cliente"]])
```

```python
#Aqui se tranforma todo de una vez
#arreglo de tuplos nombre, tipo y columna
transformer = ColumnTransformer(
    [
        ('onehot',OneHotEncoder(sparse=False),[0]),
        ('ordinal',OrdinalEncoder(categories=[["malo","regular","bueno"]]),[1])
    ]
)
```

```python
transformer.fit(Df)
```

```python
transformer.transform(Df)
```
