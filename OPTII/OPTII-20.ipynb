```python
import random

gridSize = 5
minNum = 1
maxNum = 50
cards = 40

for h in range(cards):
    card = []
    randRange = range(minNum, maxNum)
    card = random.sample(randRange, gridSize * gridSize)
    for i in range(gridSize):
        string = ""
        for j in range(gridSize):
            string +=  str(card[i + j * gridSize]) + "\t"
        print(string)

    print("\n")
```
