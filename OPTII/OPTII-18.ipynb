```python
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
x = 10
serie = pd.Series([10])

for i in range(0, 100):
  y = random.randrange(2)
  if(y == 1):
    x = x-1
    serie = pd.concat([serie, pd.Series([x])], ignore_index=True)
  else:
    x = x+1
    serie = pd.concat([serie, pd.Series([x])], ignore_index=True)
```

```python
df=pd.DataFrame(serie)
df.plot()
```

```python
dfII = pd.DataFrame()
ts = df[0]

dfII.index = ts.index
dfII["t"] = ts.values
dfII["t-1"] = ts.shift(1)
dfII["t-2"] = ts.shift(2)
dfII["t-3"] = ts.shift(3)
dfII["t-4"] = ts.shift(4)
dfII["t-5"] = ts.shift(5)
dfII["t-6"] = ts.shift(6)
dfII["t-7"] = ts.shift(7)
```

```python
dfII
```

```python
dfIII = dfII.corr()
dfIII["t"].plot()
```
